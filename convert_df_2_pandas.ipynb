```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("convert_df_2_pandas").getOrCreate()
```

```python
rdd = spark.sparkContext.parallelize([("alana",1),("ahan",2)])
schema = ("name","id")
df =rdd.toDF(schema)
pd = df.toPandas()
print(pd)
```
